```python

##########데이터 전처리
search_df = train_df[(train_df['LCD']==0) & (train_df['LED']==0) & (train_df['OLED']==0) & (train_df['QLED']==0)]
train_df = train_df.drop(search_df.index, axis=0)
search_df = test_df[(test_df['LCD']==0) & (test_df['LED']==0) & (test_df['OLED']==0) & (test_df['QLED']==0)]
test_df = test_df.drop(search_df.index, axis=0)

print(train_df.info())
```

```python
train_df.describe()
```

```python

x_train = train_df.drop(['ID', '판매가'], axis=1)   # ID는 X변수에 해당되지 않음, 판매가는 Y값으로 할것
x_test = test_df.drop(['ID', '판매가'], axis=1)
y_train = train_df['판매가']
y_test = test_df['판매가']
```

```python
# (6636, 10) 과 (6636,) 의 x값과 y값의 행 갯수(자료수)가 일치해야 함
# train 값과 test값은 열갯수(변수)가 맞아야함
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

# x의 shape(6636, 10) 은 모델식에 입력되는 10개의 x변수가 있음
# 딥러닝에서는 입력x의 갯수를 사용자가 설정해야 하는데, 컴퓨터가 계산하는 shape이랑 다르면 에러발생
x_train.shape[1]
```

```python
type(x_train)
```

```python
# x_train.values 와 같음, 머신러닝&딥러닝에서는 array만 처리함, 자료분석을 위해 넘파이로 타입 변경
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
##########모델 생성
# 선형회귀
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
```

```python
##########모델 학습

model_lr.fit(x_train, y_train)
```

```python
##########모델 검증

print(model_lr.score(x_test, y_test)) #0.5462414358589345
```

## [라쏘 모델](https://wikidocs.net/39710)

```python
# 여러개의 x변수중 영향력이 작은(회귀계수값이 작은)변수는 취급하지 않음
# 라쏘와 리지모델에서 alpha값은 규제정도를 나타냄, alpha값을 높이면 규제를 강하게 함.
from sklearn.linear_model import Lasso

alphaList=[1.0, 0.1, 100]

for alphavalue in alphaList:
    model_lasso = Lasso(alpha=alphavalue)
    model_lasso.fit(x_train, y_train)
    print(model_lasso.score(x_test, y_test))
```

## [리지모델](https://wikidocs.net/39693)

```python
# 여러개의 x변수중 영향력이 작은(회귀계수값이 작은)변수는 취급하지 않음
from sklearn.linear_model import Ridge

alphaList=[1.0, 0.1, 100]

for alphavalue in alphaList:
    model_ridge = Ridge(alpha=alphavalue)
    model_ridge.fit(x_train, y_train)
    print(model_ridge.score(x_test, y_test))
```
