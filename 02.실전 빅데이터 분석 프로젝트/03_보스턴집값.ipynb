## crim/rm/lstat 세게의 변수로 각각 단순 선형 회귀 분석하기

- CRIM: 범죄율
- RM: 주택당 방 수
- RM: 주택당 방 수

```python
target = boston[['Target']]
# boston_target
crim=boston[['CRIM']]
rm=boston[['RM']]
lstat=boston['LSTAT']
```

## target ~ crim 선형회귀분석

```python
crim1 = sm.add_constant(crim, has_constant='add')
```

```python
model1 = sm.OLS(target,crim1)
fitted_model1=model1.fit()
```
